# Visualizing Lending Club Data

```python
%matplotlib inline
```

# Import modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

# Read in Loans data and clean

```python
loansData = pd.read_csv('https://github.com/Thinkful-Ed/curric-data-001-data-sets/raw/master/loans/loansData.csv')
loansData.dropna(inplace=True) # drops null values
```

# Boxplot of Amount Requested from Lending Club Data

```python
loansData.boxplot(column='Amount.Requested')
plt.show()
```

## The boxplot for the amount requested shows a range of a little over 0 dollars to 35,000 dollars, with a median requested amount of 10,000. The long whisker representing the top 25th percent of values suggests that the distribution may be right-skewed, although a histrogram would be a better visualization of the underlying distribution. It appears there are a few outliers on the upper end; more evidence that suggests a skewed distribution. This boxplot for the amount requested is quite similar to the amount that is ultimately funded in terms of general shape and pattern.  

# Histogram of Amount Requested from Lending Club Data

```python
loansData.hist(column='Amount.Requested')
plt.show()
```

## The histogram for the amount requested confirms that the distribution has a long right tail. While a boxplot gives a better feeling for the median value, the histogram gives a better idea of which values occur the most. As seen in the preceding histogram for the amount requested, the most frequently requested loan amount was for 10,000 dollars. Again, when compared to the histogram for the amount funded, the amount requested is very similar in shape. 

# QQ Plot of Amount Requested from Lending Club Data

```python
plt.figure()
graph = stats.probplot(loansData['Amount.Requested'], dist='norm', plot=plt)
plt.show()
```

# The qq plot shows that the distribution of the amount requested is not normally distributed, and appears quite similar to the funding amount in the same dataset.
